```python
# Import Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

# Import API key
from api_keys import api_key

# Incorporated Citipy to Determine City Based on Latitude and Longitude
from citipy import citipy
```

```python
# Range of Latitudes and Longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Empty List for Holding lat_lngs and Cities
lat_lngs = []
cities = []

# Create a Set of Random lat and lng Combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify Nearest City for Each lat, lng Combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the City is Unique, Add it to the Cities List
    if city not in cities:
        cities.append(city)

# Print City Count to Confirm Count
len(cities)
```

```python
# Set Base URL
url = "http://api.openweathermap.org/data/2.5/weather?"

# Set Units
units = "imperial"

#Set Query URL
query_url = f"{url}appid={api_key}&units={units}&q="

# Use Requests to Make API Call for First City and Convert Response to JSON
response = requests.get(query_url + city).json()

# Print JSON Response
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
#Set-up Empty Lists to Hold Response Data
city_name = []
country = []
date = []
cloudiness = []
humidity = []
latitude = []
longitude = []
max_temp = []
wind_speed = []

# Set Processing Record Counter
processing_record = 1

# Print Statement for Beginning of Response Log
print("Beginning Data Retrieval")
print("----------------------------------------")

# Set For Loop to Iterate Through Cities List and Make an API Call for the Given Data for Each City
for city in cities:
    
    # Set Try and Except Block to Avoid Errors While Iterating Through Cities List
    try:
        
        # Use Requests to Make API Call for Each City and Convert Response to JSON
        response = requests.get(query_url + city).json()
        
        # Append Each Empty List With Given Data For Each City
        city_name.append(response["name"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        cloudiness.append(response["clouds"]["all"])
        humidity.append(response["main"]["humidity"])
        latitude.append(response["coord"]["lat"])
        longitude.append(response["coord"]["lon"])
        max_temp.append(response["main"]["temp_max"])
        wind_speed.append(response["wind"]["speed"])
        
        # Set City Name for Response Log
        city_name_record = response["name"]
        
        # Print Processing Record Response Log for Each City
        print(f"Processing Record {processing_record} of Set 1 | {city_name_record}")
        
        # Set Processing Record to Increase By One Each Loop
        processing_record += 1
        
    except (KeyError):
        
        print("City not found. Skipping...")
        
# Print Statement for End of Response Log
print("----------------------------------------")
print("Data Retrieval Complete")
print("----------------------------------------")
```

```python
# Create a New DataFrame from Response Data
weather_data = pd.DataFrame({
    
    "City": city_name,
    "Country": country,
    "Date": date,
    "Latitude": latitude,
    "Longitude": longitude,
    "Max Temp": max_temp,
    "Humidity": humidity,
    "Cloudiness": cloudiness,
    "Wind Speed": wind_speed})

# Confirm that There Are 500+ Cities in the Newly Created DataFrame
weather_data.count()
```
